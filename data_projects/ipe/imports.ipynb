## Importação separada devido a formatação do arquivo xlsx
    - 1º Passo, colunas não separadas

```python
df1 = pd.read_excel("Dados.xlsx", header=1, usecols="B:E,L")
df1.head()
```

    - 2º Passo, coluna hobbies

```python
df_hobbies = pd.read_excel("Dados.xlsx", header=1, usecols="F:H")
df_hobbies.head()
```

```python
df_hobbies.rename(columns={
    "Hobbies": "hobby_1",
    "Unnamed: 6": "hobby_2",
    "Unnamed: 7": "hobby_3"
}, inplace=True)
df_hobbies.head()
```

    - 3º Passo, coluna animal de estimação (nulls tratados como -)

```python
df_animais = pd.read_excel("Dados.xlsx", header=1, usecols="I:J", na_values="-")
df_animais.head()
```

```python
df_animais.rename(columns={
    "Animal de estimação": "animal_1",
    "Unnamed: 9": "animal_2"
}, inplace=True)
df_animais.head()
```

# Concatenando os 3 DFs em um

```python
df_final = pd.concat([df1, df_animais, df_hobbies], axis=1)
df_final.head()
```
